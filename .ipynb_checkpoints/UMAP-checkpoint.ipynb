```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
import umap
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
print(X.shape)
X[:10]
```

```python
y = iris.target
print(y.shape)
y
```

```python
np.unique(y, return_counts=True)
```

```python
nx,ny = 9,1
fig, ax = plt.subplots(ny,nx,figsize=(nx*4,ny*3))

for i,n_neighbors in enumerate([2,3,4,5,6,7,8,9,10]):
    embed = umap.UMAP(n_components = 3, n_neighbors=n_neighbors, min_dist=0.3, metric="correlation").fit_transform(X)

    ax[i].plot(embed[:50,0], embed[:50,1], "o")
    ax[i].plot(embed[50:100,0], embed[50:100,1], "*")
    ax[i].plot(embed[100:,0], embed[100:,1], "^")

    ax[i].set_xlabel("UMAP 1")
    ax[i].set_ylabel("UMAP 2")
    ax[i].set_title("n_neighbors: %s"%n_neighbors)
```
