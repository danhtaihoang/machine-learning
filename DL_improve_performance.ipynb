## Reduce over fitting with dropout

```python
# add a new layer prior to the layer that we want to have input connections dropped-out.
from sklearn.datasets import make_classification
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout

# plotting a model
from tensorflow.keras.utils import plot_model
```

```python
# create the dataset
X, y = make_classification(n_samples=1000, n_classes=2, random_state=1)
```

```python
# determine the number of input features
n_features = X.shape[1]
print(n_features)
```

```python
# define model
model = Sequential()
model.add(Dense(10, activation='relu', kernel_initializer='he_normal', input_shape=(n_features,)))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
# fit the model
model.fit(X, y, epochs=100, batch_size=32, verbose=0)
```

```python
plot_model(model, 'model.png', show_shapes=True)
```

## Accelerate Training With Batch Normalization

```python
# example of using batch normalization
from sklearn.datasets import make_classification
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import BatchNormalization

# plotting a model
from tensorflow.keras.utils import plot_model
```

```python
# create the dataset
X, y = make_classification(n_samples=1000, n_classes=2, random_state=1)
```

```python
# determine the number of input features
n_features = X.shape[1]
# define model
model = Sequential()
model.add(Dense(10, activation='relu', kernel_initializer='he_normal', input_shape=(n_features,)))
model.add(BatchNormalization())
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
# fit the model
model.fit(X, y, epochs=100, batch_size=32, verbose=0)
```

```python
plot_model(model, 'model.png', show_shapes=True)
```

## Early Stopping

```python
from sklearn.datasets import make_classification
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# create the dataset
X, y = make_classification(n_samples=1000, n_classes=2, random_state=1)
```

```python
# determine the number of input features
n_features = X.shape[1]
# define model
model = Sequential()
model.add(Dense(10, activation='relu', kernel_initializer='he_normal', input_shape=(n_features,)))
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
# configure early stopping
es = EarlyStopping(monitor='val_loss', patience=5)
```

```python
# fit the model
history = model.fit(X, y, epochs=200, batch_size=32, verbose=0, validation_split=0.3, callbacks=[es])
```
